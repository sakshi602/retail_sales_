# COUNT & PREVIEW OF SALES DATE

```python
missing_count = sales['SalesDate'].isnull().sum()
print(f"Number of missing SalesDate values: {missing_count}")


print(sales[sales['SalesDate'].isnull()].head())
```

# Printing Statistics

```python
# Total number of unique sales transactions
total_transactions = sales['SalesID'].nunique()
print("Total Transactions:", total_transactions)

# Total revenue (sum of TotalPrice)
total_revenue = sales['TotalPrice'].sum()
print("Total Revenue:", total_revenue)

# Average discount given
average_discount = sales['Discount'].mean()
print("Average Discount Given:", average_discount)

# Quantity statistics
min_quantity = sales['Quantity'].min()
max_quantity = sales['Quantity'].max()
mean_quantity = sales['Quantity'].mean()

print("Quantity - Min:", min_quantity)
print("Quantity - Max:", max_quantity)
print("Quantity - Mean:", mean_quantity)
```

# Convert 'SalesDate' to datetime format

```python

sales['SalesDate'] = pd.to_datetime(sales['SalesDate'])

# Extract additional temporal features
sales['Year'] = sales['SalesDate'].dt.year
sales['Month'] = sales['SalesDate'].dt.month
sales['DayOfWeek'] = sales['SalesDate'].dt.dayofweek  # Monday=0, Sunday=6
sales['WeekdayName'] = sales['SalesDate'].dt.day_name()
sales['YearMonth'] = sales['SalesDate'].dt.to_period('M')
print(sales['SalesDate'].head())
print(sales['SalesDate'].tail())
```

# STATISTICS OF SALES TABLE

```python
# Sales Table
print("="*30 + " SALES TABLE " + "="*30)
print(sales.describe(include='all').round(2))
```

# STATISTICS OF Product TABLE

```python
# Product Table
print("="*30 + " product TABLE " + "="*30)
print(products.describe(include='all').round(2))
```

# Creating new Column in sales as Revenue

```python
# Create a lookup dictionary from products
price_lookup = products.set_index('ProductID')['Price']

# Map the prices into the sales DataFrame
sales['Price'] = sales['ProductID'].map(price_lookup)

# Now calculate revenue
sales['Revenue'] = sales['Quantity'] * sales['Price'] * (1 - sales['Discount'])
print(sales.head())
print (sales.tail())
```

# TOP 10 PRODUCTS BY REVENUE

```python
top_products = sales.groupby('ProductID')['Revenue'].sum().sort_values(ascending=False).head(10)
top_products = top_products.reset_index()
top_products = top_products.merge(products[['ProductID', 'ProductName']], on='ProductID', how='left')


# Plotting
plt.figure(figsize=(10,6))
plt.barh(top_products['ProductName'], top_products['Revenue'], color='skyblue')
plt.xlabel('Total Revenue')
plt.title('Top 10 Products by Revenue')
plt.gca().invert_yaxis()  # Highest at the top
plt.tight_layout()
plt.show()
```

# TOP 10 CUSTOMERS BY REVENUE

```python
top_customers = sales.groupby('CustomerID')['Revenue'].sum().sort_values(ascending=False).head(10)

# Plot
plt.figure(figsize=(10,6))
top_customers.plot(kind='bar', color='orange')
plt.title('Top 10 Customers by Revenue')
plt.xlabel('Customer ID')
plt.ylabel('Total Revenue')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Step 1: Merge sales with customers to get CityID
sales_geo = sales.merge(customers[['CustomerID', 'CityID']], on='CustomerID', how='left')

# Step 2: Merge with cities to get CityName and CountryID
sales_geo = sales_geo.merge(cities[['CityID', 'CityName', 'CountryID']], on='CityID', how='left')

# Step 3: Merge with countries to get CountryName
sales_geo = sales_geo.merge(countries[['CountryID', 'CountryName']], on='CountryID', how='left')
```

# TOP 10 CITIES BY REVENUE

```python
top_cities = sales_geo.groupby('CityName')['Revenue'].sum().sort_values(ascending=False).head(10)

top_cities.plot(kind='barh', color='orange', figsize=(10,6))
plt.xlabel('Total Revenue')
plt.title('Top 10 Cities by Revenue')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
